## Quick Reference

| Method | Use Case | Example |
|--------|----------|---------|
| `df.col` / `df['col']` | Select a single column | `reviews.points` |
| `df.iloc[r, c]` | Integer-position based (0-indexed, end **excluded**) | `reviews.iloc[0]` |
| `df.loc[r, c]` | Label-based (end **included**) | `reviews.loc[0, 'country']` |
| `df.loc[mask]` | Boolean filtering | `reviews.loc[reviews.points >= 95]` |

> **iloc vs loc gotcha:** `iloc[0:10]` returns 10 rows; `loc[0:10]` returns 11 rows (inclusive end).
